```python
import codecademylib
import pandas as pd

ad_clicks = pd.read_csv('ad_clicks.csv')
#1
# Examining few rows of ad_clicks:
print(ad_clicks.head(10)) 

#2
# Number of views from each ad platform:
utm_source_count = ad_clicks.groupby("utm_source").user_id.count().reset_index()
utm_source_count = utm_source_count.rename(columns={"user_id":"counts"})
print(utm_source_count)

#3
# Creating a new column 'is_click':
ad_clicks['is_click'] = ~ad_clicks.ad_click_timestamp.isnull()
print(ad_clicks.head(10))

#4
# Percentage of people who clicked on ads from each utm_source.
clicks_by_source = ad_clicks.groupby(["utm_source", "is_click"]).user_id.count().reset_index()

#5
# Pivoting clicks_by_source data:
clicks_pivot = clicks_by_source.pivot(columns="is_click", index="utm_source", values="user_id").reset_index()

#6 
# Creating new column in clicks_pivot:
clicks_pivot["percent_clicked"] = clicks_pivot[True] / (clicks_pivot[True] + clicks_pivot[False])
print(clicks_pivot)

#7
# Analyzing an A/B Test
experimental_group_count = ad_clicks.groupby("experimental_group").user_id.count().reset_index()
experimental_group_count = experimental_group_count.rename(columns={"user_id":"counts"})
print(experimental_group_count)

#8
# Percentage of users clicked on Ad A or Ad B
clicks_by_ad = ad_clicks.groupby(["experimental_group", "is_click"]).user_id.count().reset_index()

clicks_pivot = clicks_by_ad.pivot(columns="is_click", index="experimental_group", values="user_id").reset_index()

clicks_pivot["percent_clicked"] = clicks_pivot[True] / (clicks_pivot[True] + clicks_pivot[False])
print(clicks_pivot)

#9
a_clicks = ad_clicks[ad_clicks.experimental_group == "A"]

b_clicks = ad_clicks[ad_clicks.experimental_group == "B"]

#10
# Percentage of users who clicked on ad A by day:
clicks_by_ad_day = a_clicks.groupby(["day", "is_click"]).user_id.count().reset_index()

clicks_pivot_a = clicks_by_ad_day.pivot(columns="is_click", index="day", values="user_id").reset_index()

clicks_pivot_a["percent_clicked"] = clicks_pivot_a[True] / (clicks_pivot_a[True] + clicks_pivot_a[False])
clicks_pivot_a = clicks_pivot_a.rename(columns={"day":"day (Group A)"})
print(clicks_pivot_a)

# Percentage of users who clicked on ad A by day:
clicks_by_ad_day = b_clicks.groupby(["day", "is_click"]).user_id.count().reset_index()

clicks_pivot_b = clicks_by_ad_day.pivot(columns="is_click", index="day", values="user_id").reset_index()

clicks_pivot_b["percent_clicked"] = clicks_pivot_b[True] / (clicks_pivot_b[True] + clicks_pivot_b[False])
clicks_pivot_b = clicks_pivot_b.rename(columns={"day":"day (Group B)"})
print(clicks_pivot_b)
print("")
#Extra
# Average percent_clicked for Ad A
average_a = clicks_pivot_a.percent_clicked.mean()
print(round(float(average_a) * 100), "%")

# Average percent_clicked for Ad A
average_b = clicks_pivot_b.percent_clicked.mean()
print(round(float(average_b) * 100), "%")

# Base on the average percent_clicked for each Ad I will recommend Ad A.
```

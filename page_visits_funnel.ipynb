```python
import codecademylib
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])

# Inspecting all DataFrames:
print(visits.head(10))
print(cart.head(10))
print(checkout.head(10))
print(purchase.head(10))


# Combining visits and cart using left merge:
visits_cart = visits.merge(cart, how="left")
print(visits_cart)

# Finding the length of visits_cart_left:
print(len(visits_cart))

#Number of null row for the cart_time column:
null_count = visits_cart[visits_cart.cart_time.isnull()]
print(len(null_count))

# Percentage of user who ended up not placing a t-shirt in their cart:
percent_cart_null = (len(null_count) / float(len(visits_cart))) * 100
print("Percentage of users who did not place a t-shirt in their chat: " + str(percent_cart_null) + "%")

# Left merge for cart and checkout:
cart_checkout = cart.merge(checkout, how="left")

null_count = cart_checkout[cart_checkout.checkout_time.isnull()]
print(len(null_count))

percent_checkout_null = (len(null_count) / float(len(cart_checkout))) * 100
print("Percentage of users who did not proceed to checkout: " + str(percent_checkout_null) + "%")

# Merging all DataFrames:
all_data = visits.merge(cart, how="left").merge(checkout, how="left").merge(purchase, how="left")
print(all_data.head())

# Percentage of users proceeded to checkout, but did not purpase:
checkout_purchase = checkout.merge(purchase, how="left")

null_count = checkout_purchase[checkout_purchase.purchase_time.isnull()]
print(len(null_count))

percent_purchase_null = (len(null_count) / float(len(checkout_purchase))) * 100
print("Percentage of users who proceeded to checkout but did not purchased: " + str(percent_purchase_null) + "%")

# Average time from initial visit to final purchase:
all_data['time_to_purchase'] = \
    all_data.purchase_time - \
    all_data.visit_time

print(all_data.time_to_purchase.head())
print(all_data.time_to_purchase.mean())
```
